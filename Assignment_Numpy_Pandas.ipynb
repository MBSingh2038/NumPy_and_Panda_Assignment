```python
#Ans: 2 ---- Finding the dot product of two arrays

a = np.random.randint(2, 10, (3, 4))
print("First Matrix:- \n")
print(a)

b = np.random.randint(2, 10, (4,3))
print("\nSecond Matrix:- \n")
print(b)

#finding dot product of two arrays
result=np.dot(a,b)

print("\nThe resultant array is:- \n")
print(result)
```
